Word Embedding Techniques using Embedding Layer in Keras

```python
### Libraries USed Tensorflow> 2.0  and keras
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
##vocabulary size
voc_size=500 #larger the size the the features but processing time will be more
```

# One Hot Representation

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
#this will return the value of indices of words from sent in the voc size (here, 10000) 
```

# Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## pre padding
sent_length=8 #set the maximum sentence length so that the dimension/shape of the onehot_repr becomes equal i.e of 8 instead of varying size like 4,5 etc.
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) #padding=pre means add 0s before the nos.
print(embedded_docs) 
```

```python
 dim=10 #feature representation dimension=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```
